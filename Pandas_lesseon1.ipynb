```python
## Adicionar 1 coluna
# Apartir de uma coluna que existe
vendas_df['Comissso'] = vendas_df['Valor Final']*0.05
display(vendas_df)
```

```python
#Criar uma coluna com valor padro 
vendas_df.loc[:,"Imposto"] = 0
display(vendas_df)
```

```python
#Adicionar uma linha
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
display(vendas_dez_df)
```

```python
#adicionar uma base de dados ao final de uma 
vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

```python
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

```python
#deletar linhas e colunas comppletamente vazias
vendas_df = vendas_df.dropna(how='all',axis=1)
```

```python
#deletar linhas que possuem pelo mesno 1 valor vazios
vendas_df = vendas_df.dropna()
```

```python
#prencher valores vazios 
#prencher com a mdia da coluna
vendas_df['Comissso'] = vendas_df['Comissso'].fillna(vendas_df['Comissso'].mean())
display(vendas_df)
```

```python
#preencher com o ultimo valor
vendas_df = vendas_df.ffill()
```

```python
#Ele conta quanta transaoes foram feitas por loja (value counts)
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
#Group by
fauturamento_produto = vendas_df[['Produto','Valor Final']].groupby('Produto').sum()
display(fauturamento_produto)
```

```python
gerentes_df = pd.read_excel("Gerentes.xlsx")
display(gerentes_df)
```

```python
#Mesclar
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
